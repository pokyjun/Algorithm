```python
#   
canvas_size = 100
paper_size = 10

#  
canvas = [[0] * canvas_size for _ in range(canvas_size)]

#  
num_papers = int(input())

#    .
for i in range(num_papers):
    #   
    x, y = map(int, input().split())
    
    #    1 .
    for j in range(x, x + paper_size):
        for k in range(y, y + paper_size):
            canvas[j][k] += 1

#     .
total_area = 0
for i in range(canvas_size):
    for j in range(canvas_size):
        if canvas[i][j] > 0:
            total_area += 1

#  
print(total_area)

'''
 
 3
 3 7
15 7
 5 2
'''
```
